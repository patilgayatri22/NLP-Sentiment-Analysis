```python
import pandas as pd
from pathlib import Path

# --- Config: update these paths if your files live elsewhere (e.g., in Drive) ---
train_path = Path("train.csv")
test_path  = Path("test.csv")

# --- Column names in your raw files (no header present) ---
raw_cols = ["label", "review_title", "review_text"]

# --- Read CSVs (no header in your files) ---
def load_raw(csv_path: Path) -> pd.DataFrame:
    df = pd.read_csv(
        csv_path,
        header=None,                # because your files have no header row
        names=raw_cols,
        encoding="utf-8",
        on_bad_lines="skip"         # skip any malformed lines
    )
    # Basic cleanup
    df["review_title"] = df["review_title"].astype(str).str.strip()
    df["review_text"]  = df["review_text"].astype(str).str.strip()
    return df

train_df = load_raw(train_path)
test_df  = load_raw(test_path)

# --- Map numeric labels to human-friendly text ---
# In your data: 1 = negative (1–2 stars), 2 = positive (4–5 stars)
label_map = {1: "negative", 2: "positive"}

# ensure label is int (sometimes it loads as str)
for df in (train_df, test_df):
    df["label"] = pd.to_numeric(df["label"], errors="coerce").astype("Int64")
    # keep only the expected label values 1 or 2
    df = df[df["label"].isin([1, 2])]
    # optionally, a combined text column you might use later for modeling
    df["text"] = (df["review_title"].fillna("") + ". " + df["review_text"].fillna("")).str.strip()

# Reassign (after filtering step above)
train_df = train_df[train_df["label"].isin([1, 2])].copy()
test_df  = test_df[test_df["label"].isin([1, 2])].copy()
train_df["text"] = (train_df["review_title"].fillna("") + ". " + train_df["review_text"].fillna("")).str.strip()
test_df["text"]  = (test_df["review_title"].fillna("") + ". " + test_df["review_text"].fillna("")).str.strip()

# --- Quick overviews ---
def overview(df: pd.DataFrame, name: str):
    print(f"\n===== {name} OVERVIEW =====")
    print(f"Shape: {df.shape}")
    print("\nSample rows:")
    display(df[["label", "review_title", "review_text"]].head(10))

overview(train_df, "TRAIN")
overview(test_df,  "TEST")

# --- (Optional) Save labeled versions for later steps ---
train_out = train_path.with_name("train_labeled.csv")
test_out  = test_path.with_name("test_labeled.csv")
train_df.to_csv(train_out, index=False)
test_df.to_csv(test_out, index=False)
print(f"\nSaved:\n- {train_out}\n- {test_out}")
```

```python
import re
import pandas as pd
import numpy as np
from pathlib import Path
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

import spacy
from spacy.lang.en.stop_words import STOP_WORDS as SPACY_STOPWORDS

# Load spaCy English model (small = fast + lemmatizer)
import sys, subprocess
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    subprocess.run([sys.executable, "-m", "spacy", "download", "en_core_web_sm"], check=True)
    nlp = spacy.load("en_core_web_sm")
```

```python

def load_raw(csv_path: Path) -> pd.DataFrame:
    raw_cols = ["label", "review_title", "review_text"]  # as per your data
    df = pd.read_csv(csv_path, header=None, names=raw_cols, encoding="utf-8", on_bad_lines="skip")
    # Coerce types & strip
    df["label"] = pd.to_numeric(df["label"], errors="coerce").astype("Int64")
    df["review_title"] = df["review_title"].astype(str).str.strip()
    df["review_text"]  = df["review_text"].astype(str).str.strip()
    return df



# Ensure only label 1/2; map to text; and create a combined text column.
label_map = {1: "negative", 2: "positive"}

def standardize(df: pd.DataFrame) -> pd.DataFrame:
    df = df[df["label"].isin([1, 2])].copy()
    df["label_text"] = df["label"].map(label_map)
    if "text" not in df.columns:
        df["text"] = (df["review_title"].fillna("") + ". " + df["review_text"].fillna("")).str.strip()
    return df

train_df = standardize(train_df)
test_df  = standardize(test_df)

print("Train shape:", train_df.shape, "| Test shape:", test_df.shape)
```

```python
# --- Handle missing/malformed and duplicates ---
def clean_missing_and_dupes(df: pd.DataFrame) -> pd.DataFrame:
    # Drop rows with missing label or text
    df = df.dropna(subset=["label", "text"]).copy()
    # Remove rows where text is empty or too short after stripping
    df["text"] = df["text"].astype(str).str.strip()
    df = df[df["text"].str.len() > 0]
    # Drop exact duplicate (label, text) pairs
    before = len(df)
    df = df.drop_duplicates(subset=["label", "text"])
    dropped = before - len(df)
    if dropped:
        print(f"Dropped {dropped} duplicate rows.")
    return df

train_df = clean_missing_and_dupes(train_df)
test_df  = clean_missing_and_dupes(test_df)

# --- Basic stats: class distribution & balance ---
def class_report(df: pd.DataFrame, name: str):
    print(f"\n===== {name} CLASS DISTRIBUTION =====")
    counts = df["label_text"].value_counts().rename_axis("class").to_frame("count")
    counts["proportion"] = (counts["count"] / counts["count"].sum()).round(3)
    print(counts)
    imbalance_ratio = counts["count"].max() / counts["count"].min() if len(counts) == 2 else np.nan
    print(f"Imbalance ratio (majority/minority): {imbalance_ratio:.2f}")

class_report(train_df, "TRAIN")
class_report(test_df,  "TEST")

# --- Review length distribution (characters & tokens) ---
def add_length_metrics(df: pd.DataFrame) -> pd.DataFrame:
    # quick whitespace token count (pre-tokenization)
    df["n_chars"] = df["text"].str.len()
    df["n_words"] = df["text"].str.split().str.len()
    return df

train_df = add_length_metrics(train_df)
test_df  = add_length_metrics(test_df)

print("\n===== TRAIN LENGTH STATS =====")
print(train_df[["n_chars", "n_words"]].describe().round(2))

print("\n===== TEST LENGTH STATS =====")
print(test_df[["n_chars", "n_words"]].describe().round(2))

```

```python

import re, numpy as np, pandas as pd
from tqdm.auto import tqdm
import nltk
from nltk.stem import SnowballStemmer


mode = "ml_stem"     
batch_rows = 200_000 

# Negations kept to preserve sentiment cues
NEGATIONS = {"no", "not", "nor", "never", "n't"}

# spaCy's stopwords if available from earlier, else fall back to a small list
try:
    from spacy.lang.en.stop_words import STOP_WORDS as SPACY_STOPWORDS
    base_stop = set(SPACY_STOPWORDS)
except Exception:
    base_stop = {
        "a","an","the","and","or","but","if","while","with","for","from","to","in","on","at","by",
        "is","am","are","was","were","be","been","being","of","as","it","this","that","these","those",
        "i","you","he","she","we","they","my","your","his","her","our","their","me","him","them","so",
        "do","does","did","doing","can","could","should","would","will","just","than","then","there","here"
    }

stopwords = base_stop.difference(NEGATIONS)

stemmer = SnowballStemmer("english")


# Regexes (compiled once)
url_re   = re.compile(r"https?://\S+|www\.\S+")
html_re  = re.compile(r"<.*?>")
mention_re = re.compile(r"[@#]\w+")
# Keep letters + whitespace only (after basic cleaning)
non_letter_re = re.compile(r"[^a-z\s]+")
# Quick tokenizer to alphabetic tokens
token_re = re.compile(r"[a-z]+")

def normalize_minimal(s: str) -> str:
    """Super fast cleanup: lowercase, strip URLs/HTML/mentions, collapse whitespace."""
    s = s.lower()
    s = url_re.sub(" ", s)
    s = html_re.sub(" ", s)
    s = mention_re.sub(" ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def process_ml_stem(s: str) -> str:
    """
    Classic ML cleaning: minimal normalize -> letters only -> remove stopwords (keep negations) -> stem.
    Fast and robust for TF-IDF + linear models.
    """
    s = normalize_minimal(s)
    s = non_letter_re.sub(" ", s)        # remove punctuation/numbers
    toks = token_re.findall(s)            # quick tokenization
    toks = [t for t in toks if t not in stopwords and len(t) > 1]
    toks = [stemmer.stem(t) for t in toks]
    return " ".join(toks)

def process_light(s: str) -> str:
    """
    Light cleaning for transformer embeddings (SBERT): keep words mostly intact.
    """
    s = normalize_minimal(s)
    # Optional: very gentle punctuation trim that keeps basic sentence structure
    return s

def preprocess_series(series: pd.Series, mode: str, batch_rows: int = 200_000) -> pd.Series:
    out = []
    n = len(series)
    fn = process_ml_stem if mode == "ml_stem" else process_light

    for start in tqdm(range(0, n, batch_rows), desc=f"Preprocessing ({mode})", unit="rows"):
        chunk = series.iloc[start:start+batch_rows].tolist()
        out.extend([fn(x if isinstance(x, str) else "") for x in chunk])

    return pd.Series(out, index=series.index)


# Assumes you already have train_df/test_df with a 'text' column from earlier steps.
# If any rows became empty after cleaning, we drop them to be safe.
def apply_and_attach(df: pd.DataFrame, name: str) -> pd.DataFrame:
    cleaned = preprocess_series(df["text"], mode=mode, batch_rows=batch_rows)
    tmp = df.copy()
    tmp["text_clean"] = cleaned
    before = len(tmp)
    tmp = tmp[tmp["text_clean"].str.len() > 0]
    if len(tmp) < before:
        print(f"{name}: removed {before - len(tmp)} rows with empty cleaned text.")
    return tmp

train_clean = apply_and_attach(train_df, "TRAIN")
test_clean  = apply_and_attach(test_df,  "TEST")

# Quick sanity check
print("\nShapes:", train_clean.shape, test_clean.shape)
print("\nSample:")
print(train_clean[["text"]].head(2))
print(train_clean[["text_clean"]].head(2))
```

```python
# If you want explicit token lists (word-level), build from the cleaned text:
def tokenize_to_list(texts: pd.Series) -> list[list[str]]:
    out = []
    for doc in nlp.pipe(texts.tolist(), batch_size=512, disable=["ner"]):
        toks = [t.text for t in doc if t.text and not t.is_space]
        out.append(toks)
    return out


train_clean = train_clean.assign(tokens=train_clean["text_clean"].str.split())
test_clean  = test_clean.assign(tokens=test_clean["text_clean"].str.split())
print("Example tokens:", train_clean["tokens"].iloc[0][:20])
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import vstack

#Fit on a LARGE SAMPLE to build a stable vocab/idf
rng = np.random.RandomState(42)
sample_size = min(300_000, len(train_clean))
sample_idx = rng.choice(len(train_clean), size=sample_size, replace=False)

vectorizer = TfidfVectorizer(
    analyzer="word", ngram_range=(1,2),
    min_df=10, max_df=0.9,
    max_features=800_000,    # HARD CAP — critical
    dtype=np.float32,        # halve memory vs float64
    sublinear_tf=True,
    strip_accents="unicode",
    lowercase=False
)
vectorizer.fit(train_clean["text_clean"].iloc[sample_idx])

# Transform the FULL sets in chunks (progress + lower RAM spikes)
def transform_in_chunks(texts, vec, chunksize=200_000):
    parts = []
    for start in range(0, len(texts), chunksize):
        part = vec.transform(texts.iloc[start:start+chunksize])
        parts.append(part)
    X = vstack(parts)
    del parts
    return X

X_train_tfidf = transform_in_chunks(train_clean["text_clean"], vectorizer, chunksize=200_000)
X_test_tfidf  = transform_in_chunks(test_clean["text_clean"],  vectorizer, chunksize=200_000)

y_train = train_clean["label"].astype(int).to_numpy()
y_test  = test_clean["label"].astype(int).to_numpy()
print("TF-IDF shapes:", X_train_tfidf.shape, X_test_tfidf.shape)
```

```python


import os, math, time, numpy as np, pandas as pd, matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve

# Assuming already have train_clean, test_clean with columns:
# ['label', 'label_text', 'text', 'text_clean'] from your earlier steps.
assert {"label", "text_clean"}.issubset(train_clean.columns), "Missing columns in train_clean"
assert {"label", "text_clean"}.issubset(test_clean.columns),  "Missing columns in test_clean"

# Binary labels: 1 = negative, 2 = positive  -> map to y in {0,1}
def to_binary(y_series: pd.Series):
    y = (y_series.astype(int) == 2).astype(np.int32)  # positive=1
    return y

# -------------------
# Config you can tune
# -------------------
SEED = 42
tf.keras.utils.set_random_seed(SEED)

USE_SUBSET = False        # True to iterate faster first time; set False to train on full set
TRAIN_N    = 500_000     # how many train rows to use if USE_SUBSET=True
TEST_N     = 200_000     # how many test rows for quick eval/plots if USE_SUBSET=True

BATCH_SIZE = 2048        # large batches are fine (we average over tokens)
EPOCHS     = 25           # start small; increase if underfitting
LR         = 1e-3
HASH_BINS  = 2**20       # number of hash buckets (vocab bins)
EMBED_DIM  = 128         # embedding size
HIDDEN_UNITS = 128
DROPOUT    = 0.2

# Subset for speed (optional)
if USE_SUBSET:
    train_used = train_clean.sample(TRAIN_N, random_state=SEED)
    test_used  = test_clean.sample(TEST_N,  random_state=SEED)
else:
    train_used = train_clean
    test_used  = test_clean

print("Train used:", len(train_used), " | Test used:", len(test_used))
```

```python
# We keep text as raw strings; the model will split + hash on the fly.

def make_dataset(df: pd.DataFrame, batch_size: int, shuffle: bool) -> tf.data.Dataset:
    texts = df["text_clean"].astype(str).tolist()
    labels = to_binary(df["label"]).tolist()

    ds = tf.data.Dataset.from_tensor_slices((texts, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=min(1_000_000, len(df)), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_dataset(train_used, BATCH_SIZE, shuffle=True)
val_ds   = make_dataset(test_used.iloc[:len(test_used)//2], BATCH_SIZE, shuffle=False)   # small holdout for early stopping
test_ds  = make_dataset(test_used, BATCH_SIZE, shuffle=False)
```

```python
# Model: Input (string) -> split -> Hashing(num_bins) -> Embedding(num_bins, EMBED_DIM)
#         -> mean over tokens (EmbeddingBag) -> Dense -> Dropout -> Sigmoid

# Keras functional model that handles ragged tokens
text_in = tf.keras.Input(shape=(), dtype=tf.string, name="text")

# tokenize on whitespace (RaggedTensor of shape [batch, (tokens)])
tokens = tf.strings.split(text_in)

# hash each token to a bucket (stable, fixed-size "vocab")
hasher = tf.keras.layers.Hashing(num_bins=HASH_BINS)
token_ids = hasher(tokens)  # Ragged[int]

# embed each token id
embed = tf.keras.layers.Embedding(input_dim=HASH_BINS, output_dim=EMBED_DIM, name="embed")
embeddings = embed(token_ids)  # Ragged[batch, tokens, EMBED_DIM]

# average over token dimension (works directly on ragged tensors)
x = tf.reduce_mean(embeddings, axis=1)  # [batch, EMBED_DIM]

# optional hidden layer
x = tf.keras.layers.Dense(HIDDEN_UNITS, activation="relu")(x)
x = tf.keras.layers.Dropout(DROPOUT)(x)

out = tf.keras.layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(text_in, out)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LR),
    loss="binary_crossentropy",
    metrics=["accuracy", tf.keras.metrics.Precision(name="precision"), tf.keras.metrics.Recall(name="recall")]
)

model.summary()
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=10, restore_best_weights=True)
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# --- Plot training curves (Loss & Accuracy) ---
def plot_history(hist):
    # Loss
    plt.figure()
    plt.plot(hist.history["loss"])
    plt.plot(hist.history["val_loss"])
    plt.title("Training & Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(["train", "val"])
    plt.show()

    # Accuracy
    plt.figure()
    plt.plot(hist.history["accuracy"])
    plt.plot(hist.history["val_accuracy"])
    plt.title("Training & Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend(["train", "val"])
    plt.show()

    # Precision
    if "precision" in hist.history and "val_precision" in hist.history:
        plt.figure()
        plt.plot(hist.history["precision"])
        plt.plot(hist.history["val_precision"])
        plt.title("Training & Validation Precision")
        plt.xlabel("Epoch")
        plt.ylabel("Precision")
        plt.legend(["train", "val"])
        plt.show()

    # Recall
    if "recall" in hist.history and "val_recall" in hist.history:
        plt.figure()
        plt.plot(hist.history["recall"])
        plt.plot(hist.history["val_recall"])
        plt.title("Training & Validation Recall")
        plt.xlabel("Epoch")
        plt.ylabel("Recall")
        plt.legend(["train", "val"])
        plt.show()

plot_history(history)
```

```python
# Get predicted probabilities on the test set
y_true = []
y_prob = []
for xb, yb in test_ds:
    y_true.append(yb.numpy())
    y_prob.append(model.predict(xb, verbose=0).ravel())

y_true = np.concatenate(y_true)
y_prob = np.concatenate(y_prob)
y_pred = (y_prob >= 0.5).astype(int)

print("Evaluation at threshold 0.5")
print(classification_report(y_true, y_pred, target_names=["negative","positive"], digits=4))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

# Compute ROC & PR for curves
fpr, tpr, _ = roc_curve(y_true, y_prob)
roc_auc = auc(fpr, tpr)
prec, rec, _ = precision_recall_curve(y_true, y_prob)
```

```python


import numpy as np, pandas as pd, matplotlib.pyplot as plt, tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve


assert {"label","text_clean"}.issubset(train_clean.columns), "train_clean must have label & text_clean"
assert {"label","text_clean"}.issubset(test_clean.columns),  "test_clean must have label & text_clean"

# Binary labels: map {1->0 (negative), 2->1 (positive)}
def to_binary(y: pd.Series) -> np.ndarray:
    return (y.astype(int) == 2).astype(np.int32).to_numpy()


SEED         = 42
USE_SUBSET   = False      # Start True for a quick, stable run; set False for full dataset later
TRAIN_N      = 500_000   # used when USE_SUBSET=True
TEST_N       = 200_000   # used when USE_SUBSET=True

# Sequence / vocab
MAX_TOKENS   = 100_000   # vocabulary cap for TextVectorization
SEQ_LEN      = 200       # truncate/pad length

# Model capacity & training
EMBED_DIM    = 128
LSTM_UNITS   = 64
DROPOUT      = 0.3
BATCH_SIZE   = 1024
EPOCHS       = 50
LR           = 1e-3



tf.keras.utils.set_random_seed(SEED)

# Subset for fast iteration
if USE_SUBSET:
    train_used = train_clean.sample(TRAIN_N, random_state=SEED)
    test_used  = test_clean.sample(TEST_N,  random_state=SEED)
else:
    train_used = train_clean
    test_used  = test_clean

print("Train used:", len(train_used), "| Test used:", len(test_used))
```

```python
AUTOTUNE = tf.data.AUTOTUNE

# Create a text-only dataset for adaptation without materializing giant Python lists
def text_ds_from_df(df: pd.DataFrame, batch_size: int = 4096) -> tf.data.Dataset:
    return tf.data.Dataset.from_tensor_slices(df["text_clean"].astype(str).values).batch(batch_size)

# Build the vectorizer
text_vec = tf.keras.layers.TextVectorization(
    max_tokens=MAX_TOKENS,
    standardize="lower_and_strip_punctuation",
    split="whitespace",
    output_mode="int",
    output_sequence_length=SEQ_LEN
)

# Adapt on a large sample for speed (tune SAMPLE_ADAPT)
SAMPLE_ADAPT = min(800_000, len(train_used))
adapt_ds = text_ds_from_df(train_used.sample(SAMPLE_ADAPT, random_state=SEED))
text_vec.adapt(adapt_ds)
print("Vectorizer vocab size (cap):", text_vec.vocabulary_size())

# Build tf.data pipelines
def make_supervised_ds(df: pd.DataFrame, batch_size: int, shuffle: bool) -> tf.data.Dataset:
    x = df["text_clean"].astype(str).values
    y = to_binary(df["label"])
    ds = tf.data.Dataset.from_tensor_slices((x, y))
    if shuffle:
        ds = ds.shuffle(buffer_size=min(len(df), 1_000_000), seed=SEED, reshuffle_each_iteration=True)
    # map -> vectorize -> batch -> prefetch
    def _vectorize(x, y):
        return text_vec(x), y
    ds = ds.batch(batch_size).map(_vectorize, num_parallel_calls=AUTOTUNE).prefetch(AUTOTUNE)
    return ds

# Split out a validation slice from test for early stopping
val_frac = 0.5
val_count = int(len(test_used) * val_frac)
val_df = test_used.iloc[:val_count]
test_df_eval = test_used.iloc[val_count:]

train_ds = make_supervised_ds(train_used, BATCH_SIZE, shuffle=True)
val_ds   = make_supervised_ds(val_df,  BATCH_SIZE, shuffle=False)
test_ds  = make_supervised_ds(test_df_eval, BATCH_SIZE, shuffle=False)

print("Batches -> train:", len(list(train_ds.take(1))), "val:", len(list(val_ds.take(1))), "test:", len(list(test_ds.take(1))))
```

```python
from tensorflow import keras
from tensorflow.keras import layers

vocab_size = text_vec.vocabulary_size()

inputs = keras.Input(shape=(SEQ_LEN,), dtype=tf.int64, name="tokens")
x = layers.Embedding(input_dim=vocab_size, output_dim=EMBED_DIM, mask_zero=True, name="embed")(inputs)
x = layers.SpatialDropout1D(DROPOUT)(x)
x = layers.Bidirectional(layers.LSTM(LSTM_UNITS))(x)   # return_sequences=False -> last hidden state
x = layers.Dropout(DROPOUT)(x)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(DROPOUT)(x)
outputs = layers.Dense(1, activation="sigmoid")(x)

bilstm_model = keras.Model(inputs, outputs, name="BiLSTM_Sentiment")

bilstm_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LR),
    loss="binary_crossentropy",
    metrics=["accuracy", keras.metrics.Precision(name="precision"), keras.metrics.Recall(name="recall")]
)

bilstm_model.summary()
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=5, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=1, min_lr=1e-5, verbose=1)
]

history = bilstm_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    verbose=1,
    callbacks=callbacks
)

# Plots: Loss / Accuracy / Precision / Recall
def plot_history(hist):
    def _plot(metric, title=None):
        if metric not in hist.history: return
        plt.figure()
        plt.plot(hist.history[metric], label=f"train_{metric}")
        val_key = f"val_{metric}"
        if val_key in hist.history:
            plt.plot(hist.history[val_key], label=f"val_{metric}")
        plt.title(title or metric)
        plt.xlabel("Epoch"); plt.ylabel(metric.capitalize()); plt.legend(); plt.show()

    _plot("loss", "Training & Validation Loss")
    _plot("accuracy", "Training & Validation Accuracy")
    _plot("precision", "Training & Validation Precision")
    _plot("recall", "Training & Validation Recall")

plot_history(history)
```

```python
# Collect predictions on full test split
y_true = []
y_prob = []
for xb, yb in test_ds:
    y_true.append(yb.numpy())
    y_prob.append(bilstm_model.predict(xb, verbose=0).ravel())

y_true = np.concatenate(y_true)
y_prob = np.concatenate(y_prob)

# Default threshold 0.5
y_pred = (y_prob >= 0.5).astype(int)

print("== Classification report @ threshold 0.5 ==")
print(classification_report(y_true, y_pred, target_names=["negative","positive"], digits=4))

# Compute F1-optimal threshold on validation slice (optional, tends to improve balance)
val_true, val_prob = [], []
for xb, yb in val_ds:
    val_true.append(yb.numpy())
    val_prob.append(bilstm_model.predict(xb, verbose=0).ravel())
val_true = np.concatenate(val_true)
val_prob = np.concatenate(val_prob)

# Find threshold maximizing F1 on validation
from sklearn.metrics import f1_score
thresholds = np.linspace(0.3, 0.7, 41)
f1s = [f1_score(val_true, (val_prob >= t).astype(int)) for t in thresholds]
best_idx = int(np.argmax(f1s))
best_t = float(thresholds[best_idx])
best_f1 = float(f1s[best_idx])
print(f"Best threshold on validation: {best_t:.3f} (F1={best_f1:.4f})")

# Re-evaluate on test with best threshold
y_pred_best = (y_prob >= best_t).astype(int)
print(f"\n== Classification report @ threshold {best_t:.3f} ==")
print(classification_report(y_true, y_pred_best, target_names=["negative","positive"], digits=4))

# Confusion matrix (best threshold)
cm = confusion_matrix(y_true, y_pred_best)
print("Confusion Matrix:\n", cm)

# ROC / PR
fpr, tpr, _ = roc_curve(y_true, y_prob)
roc_auc = auc(fpr, tpr)
prec, rec, _ = precision_recall_curve(y_true, y_prob)
```

```python
# Confusion matrix heatmap
plt.figure()
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix (BiLSTM)")
plt.colorbar()
ticks = np.arange(2)
plt.xticks(ticks, ["negative","positive"], rotation=45)
plt.yticks(ticks, ["negative","positive"])
thresh = cm.max() / 2.0
for i in range(2):
    for j in range(2):
        plt.text(j, i, format(cm[i, j], "d"),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")
plt.ylabel("True label"); plt.xlabel("Predicted label")
plt.tight_layout(); plt.show()

# ROC curve
plt.figure()
plt.plot(fpr, tpr, label=f"AUC={roc_auc:.4f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.title("ROC Curve (BiLSTM)")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.legend(); plt.tight_layout(); plt.show()

# PR curve
plt.figure()
plt.plot(rec, prec)
plt.title("Precision-Recall Curve (BiLSTM)")
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.tight_layout(); plt.show()

# Predicted probability histograms by class
plt.figure()
plt.hist(y_prob[y_true==0], bins=50, alpha=0.5, label="negative")
plt.hist(y_prob[y_true==1], bins=50, alpha=0.5, label="positive")
plt.title("Predicted P(positive) Distribution by Class")
plt.xlabel("Predicted probability"); plt.ylabel("Count"); plt.legend()
plt.tight_layout(); plt.show()

# Threshold vs F1 on validation
plt.figure()
plt.plot(thresholds, f1s, marker="o")
plt.title("Threshold Tuning on Validation (F1)")
plt.xlabel("Threshold"); plt.ylabel("F1-score")
plt.tight_layout(); plt.show()
```
